```python
import pandas as pd
import numpy as np
```

```python
wk717=pd.read_csv('Week071723.csv')
wk724=pd.read_csv('Week072423.csv')
wk731=pd.read_csv('Week073123.csv')
wk807=pd.read_csv('Week080723.csv')
wk814=pd.read_csv('Week081423.csv')
wk821=pd.read_csv('Week082123.csv')
wk828=pd.read_csv('Week082823.csv')
wk904=pd.read_csv('Week090423.csv')
```

```python
columns=['user_id','country','device','device_type','language','os','paying',
     'region','gp:Channels','gp:DateCreated','gp:NumberOfChannels',
     'gp:Platform','gp:date_first_subscription_created','gp:initial_utm_campaign','gp:initial_utm_content',
     'gp:initial_utm_medium','gp:initial_utm_source','gp:onboarding_checklist','gp:purchased_a_license_atleast_once',
     'gp:referred_by_user_id'
    
    ]
```

```python
wk717=wk717.loc[:, wk717.columns.isin(columns)]
wk717['WeekActivated']='07/17/23'
wk724=wk724.loc[:, wk724.columns.isin(columns)]
wk724['WeekActivated']='07/24/23'
wk731=wk731.loc[:, wk731.columns.isin(columns)]
wk731['WeekActivated']='07/31/23'
wk807=wk807.loc[:, wk807.columns.isin(columns)]
wk807['WeekActivated']='08/07/23'
wk814=wk814.loc[:, wk814.columns.isin(columns)]
wk814['WeekActivated']='08/14/23'
wk821=wk821.loc[:, wk821.columns.isin(columns)]
wk821['WeekActivated']='08/21/23'
wk828=wk828.loc[:, wk828.columns.isin(columns)]
wk828['WeekActivated']='08/28/23'
wk904=wk904.loc[:, wk904.columns.isin(columns)]
wk904['WeekActivated']='09/04/23'


```

```python
wkall=pd.concat([wk717, wk724, wk731,wk807,wk814,wk821,wk828,wk904], axis=0, ignore_index=False)
```

```python
wkall
```

```python
activation=wkall[['user_id','WeekActivated']]
```

```python
Users=pd.DataFrame({'Users':pd.Series(wkall['user_id'].unique())})
```

```python
Weeks=pd.DataFrame({'Week':['07/17/23','07/24/23','07/31/23','08/07/23','08/14/23','08/21/23','08/28/23','09/04/23']})
```

```python
BaseDF=Weeks.merge(Users, how='cross')
```

```python
df=BaseDF.merge(activation,left_on=['Users','Week'],right_on=['user_id','WeekActivated'],how='left')
```

```python
df['Activated'] = np.where(pd.notna(df['WeekActivated']), 1, 0)
df=df.drop(['WeekActivated'], axis=1)
df
```
